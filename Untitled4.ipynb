```
import numpy as np
import pandas as pd
import tensorflow as tf
tf.__version__
```

```
dataset=pd.read_excel('Folds5x2_pp.xlsx')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
print(x,y)
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=0)
```

***BUILDING THE ANN***

```
initializANN=tf.keras.models.Sequential()
```

*** BUILDING THE INPUT LAYER AND FIRST HIDDEN LAYER***

```
initializANN.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

***ADDING SECOND LAYER AND OUTPUT LAYER***

```
initializANN.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

***OUTPUT LAYER***

```
initializANN.add(tf.keras.layers.Dense(units=1))
```

*** TRAINING THE ANN   ***

*** COMPILICOMPILING THE CODE ***

```
initializANN.compile(optimizer='adam' , loss='mean_squared_error') 
##adam is the optimizer like go from unput to output and optimize the result  and the error is the mean squared loss## 
```

***TRAININIG THE MODEL IN TRAING  SET ***

```
initializANN.fit(x_train,y_train,batch_size=32,epochs=750)
```

***PRIDICTION THE RESULT***

```
y_prediction=initializANN.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_prediction.reshape(len(y_prediction),1),y_test.reshape(len(y_test),1)),1))
```
